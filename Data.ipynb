```python
import pandas as pd

import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

Let's create the m value according to the output with the help of the numpy library

```python
m = np.arange(0, 11, dtype= "float")
c = 3* 10**8
E = m* c**2
```

PLOTTING E=mc^2

```python
plt.figure(figsize=(10, 6))
plt.plot(m, E, marker='o', linestyle='-', color='blue', label='E=mcÂ²')
plt.title('Energy (E) vs Mass (m) (mc^2)')
plt.xlabel('Mass (m) in g')
plt.ylabel('Energy (E) in Joules')
plt.grid(True)
plt.legend()
plt.show()
```

```python

# Constructing the graph using the same variables m and E
plt.figure(figsize=(10, 6))
plt.plot(m, E, marker='o', linestyle='-', label='E=mcÂ²')

plt.yscale('log')

plt.title('Energy (E) vs Mass (m)')
plt.xlabel('Mass (m) in kg')
plt.ylabel('Energy (E) in Joules (log scale)')

plt.grid(True, which="both", linestyle='--', linewidth=0.5, axis='y')
plt.legend()
plt.show()
```

```python
labels = ["1 Mo", "3 Mo", "6 Mo", "1 Yr", "2 Yr", "3 Yr", "5 Yr", "7 Yr", "10 Yr", "20 Yr", "30 Yr"]
july16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]

plt.figure(figsize=(12, 6))
plt.plot(labels, july16_2007, marker='o', linestyle='-', color='blue', label='July 16, 2007')
plt.plot(labels, july16_2020, marker='o', linestyle='-', color='green', label='July 16, 2020')

plt.xlabel('Zaman')

plt.grid(True, which="both", linestyle='--', linewidth=0.5)

plt.legend()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(10, 6))


ax1.plot(labels, july16_2007, marker='o', linestyle='-', color='blue')
ax1.set_ylabel('July 16 2007', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

ax2 = ax1.twinx()
ax2.plot(labels, july16_2020, marker='o', linestyle='-', color='red')
ax2.set_ylabel('July 16 2020', color='red')
ax2.tick_params(axis='y', labelcolor='red')
plt.show()
```
